# Context Managers

# @contextmanager

The @contextmanager decorator is used to define a context manager in a simpler way than creating a class with __enter__ and __exit__ methods. It allows you to write a generator function where the code before the yield is executed when entering the context, and the code after the yield is executed when exiting the context.

An abstract example:

```python
from contextlib import contextmanager

@contextmanager
def manage_resource(*args, **kwargs):
    # Code to acquire the resource
    resource = acquire_resource(*args, **kwargs)
    try:
        yield resource # A generator-iterator must be returned (only one value can be yielded)
    finally:
        # Code to release the resource
        release_resource(resource)
```

The function can then be used like

```python
with manage_resource("some_resource") as resource:
    # Use the resource
    pass
```


````python
from contextlib import contextmanager

# Example 1
@contextmanager
def my_context_manager():
    # Everything before the yield is executed when entering the context
    print("=" * 20 + " Example 1 Start " + "=" * 20)
    try:
        yield # This value is returned to the with statement (in this case, None)
    finally:
        # Everything after the yield is executed when exiting the context
        print("=" * 20 + " Example 1 End " + "=" * 20)

with my_context_manager():
    print("Inside the context")

# Example 2
@contextmanager
def file_manager(file_name, mode):
    print("Before yield")
    f = open(file_name, mode)
    try:
        yield f # This value is returned to the with statement's target variable (variable after the "as" keyword)
    finally:
        f.close()
        print("After yield")

from pathlib import Path
file = Path("../artifacts/contextmanager.txt")
with file_manager(file, "r") as f:
    content = f.read()
    print(content)

print("=" * 20 + " Example 2 End " + "=" * 20)
````

If an 'unhandled exception' occurs, It can be either handled or suppressed by the context manager. If the exception is re-raised explicitly, it is propagated to the outer scope. If not re-raised, the exception is suppressed.


````python
# Example 3: Unhandled Exception in Context Manager

# Example 3.1: Exception is not reraised
print("Example 3.1: Exception is not reraised")
@contextmanager
def exception_context_manager():
    print("Entering context")
    try:
        yield
    except Exception as e:
        print(f"Exception caught: {e}")
        # Exception is suppressed (not re-raised)
    finally:
        print("Exiting context")

with exception_context_manager():
    print("Inside context")
    raise ValueError("An error occurred")

# Example 3.2: Exception is reraised
print()
print("Example 3.2: Exception is reraised")
@contextmanager
def exception_context_manager_reraise():
    print("Entering context")
    try:
        yield
    except Exception as e:
        print(f"Exception caught: {e}")
        raise # Re-raise the exception to propagate it to the outer scope
    finally:
        print("Exiting context")

with exception_context_manager_reraise():
    print("Inside context")
    raise ValueError("An error occurred")
````
